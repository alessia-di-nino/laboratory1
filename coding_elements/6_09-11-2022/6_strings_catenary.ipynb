```python
# Integer numbers.

a = 1
print(type(a))

a = -1
print(type(a))


# Real numbers (aka floats).

a = 1.0
print(type(a))


# Complex numbers.

a = 1.0j
print(type(a))

a = 2.0 + 1.0j
print(type(a))


# Convert between types.

a = int(1.0)
print(a)
print(type(a))

a = int(1.9)
print(a)

#a = float(1.0J)  # error

#a = float(2.0 + 0.0J)  # error


# Strings.

a = "hello"
print(type(a))


# String formatting I.  (Old style, for shorter strings.)

# Check out: pyformat.info

s = "Hello %d World." % (1)  # integer
print(s)

s = "Hello %d World"
print(s % (1))  # substitution is delayed

s = "Hello %d %d World." % (1,2)  # several substitutions
print(s)

n = 1
s = "Hello %d World." % (n)  # can use variable
print(s)

s = "Hello %3d World." % (1)  # number of digits
print(s)

s = "Hello %03d World." % (1)  # zero padding
print(s)

s = "Hello %f World." % (1.0)  # float
print(s)

s = "Hello %8.3f World." % (1.0)  # width and decimal places
print(s)

s = "Hello %s." % ("World")  # string
print(s)

s = "Hello %10s." % ("World")  # padded string
print(s)


# String formatting II.  (Used in Prof. Baldini's lecture notes.)

x = 2
y = 1.0
s = f"Hello {x:03d} {y:8.3f} World."
print(s)


# String formatting III.  (Useful for long strings.)

s = "Hello {1:03d} {0:8.3f} {1:03d} World.".format(1.0, 2)
print(s)

s = "Hello {name} {surname}.".format(name="William", surname="Shockley")
print(s)

d = {"name": "William", "surname": "Shockley"}  # dictionary
s = "Hello {name} {surname}.".format(**d)
print(s)


# Escaping.

s = "Hello\nWorld."
print(s)

s = r"Hello\nWorld."
print(s)

s = '''This is
an extended
piece of text.
'''
print(s)


# Strings to and from files.

f = open("hello.txt", "w")  # check that the folder is correct!
f.write("Hello World.")
f.close()

f = open("hello.txt", "r")
s = f.read()  # use readlines to save lines separately as list elements
f.close()
print(s)


```

```python
# Esempio di codice per l'esperienza di laboratorio della catenaria.
# Consultare la traccia dell'esercitazione per una spiegazione completa.

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from scipy.optimize import curve_fit

# Produrre un file di testo che contenga le coordinate x e y, in pixel, di
# alcuni punti sulla fotografia della cordicella.  Una decina di punti è
# sufficiente ad ottenere un buon fit.
# Per identificare le coordinate in pixel è possibile, ad esempio, usare un
# editor di fotografie come Paint (su Windows) oppure GIMP (multipiattaforma).
# Solitamente, la posizione del cursore viene indicata come coppia di pixel
# in una barra di stato dell'interfaccia grafica.
# Notare che i pixel vengono contati da sinistra a destra e dall'alto in basso.
# Qui si assume che le colonne nel file di dati siano delimitate da virgole,
# quindi il file si dice "CSV" (comma-separated values).

dati = np.loadtxt("./catenaria.csv", delimiter=",")

# L'array dati contiene le posizioni x e y nella prima colonna (indice 0) e
# nella seconda colonna (indice 1), rispettivamente.  Per comodità, copiamo
# i valori in due array separati.  Il simbolo "due punti" serve a indicare
# tutte le righe dell'array.
# Conviene cambiare il segno ai pixel sull'asse y, in modo da ovviare al fatto
# che nelle foto vengono contati dall'alto in basso.  In questo modo, l'asse
# delle y diventa orientato dal basso in alto (come ci aspettiamo) con lo zero
# posizionato all'estremo superiore della figura.
px = dati[:,0]
py = -dati[:,1]

# Figura che mostra i dati raccolti.
fig = plt.figure(figsize=(3.0,3.0), frameon=True)
plt.axes([0.25,0.2,0.7,0.7])
plt.plot(px, py, "+r")
plt.xlim([0,800])
plt.ylim([-800,0])
plt.xlabel(r"x [px]")
plt.ylabel(r"y [py]")
plt.savefig("fit.png", dpi=300)

# Modello per il fit.
# Il primo argomento della funzione è la variabile indipendente, gli altri tre
# argomenti sono i parametri di fit.
def catenaria(x, a, c, x0):
    y = c + a * np.cosh((x - x0) / a)
    return y

# Valore di guess per il fit, che corrisponde all'ordine di grandezza atteso
# per i tre parametri di fit (a,c,x0).
# Se la foto ha una risoluzione molto alta, i pixel varieranno attorno al valore
# 1000, altrimenti attorno al valore 100.
p_guess = (100.0, 100.0, 100.0)

# La funzione curve_fit restituisce i valori ottimali dei parametri di fit
# nella variabile popt.
popt, pcov = curve_fit(catenaria, px, py, p_guess)

# Calcoliamo la funzione di fit su una griglia arbitraria di valori della
# variabile indipendente.  Ora dobbiamo passare alla funzione di fit
# i valori ottimizzati dei parametri di fit, che sono stati salvati nella
# variabile popt e sono accessibili come popt[0], popt[1], popt[2].
# Per calcolare la funzione "catenaria" su tutti i valori della variabile
# indipendente x, è possibile passare un array alla funzione (che
# automaticamente verrà eseguita per ogni elemento dell'array) oppure
# costruire un array di y iterando la variabile x su tutta la griglia di valori
# con un ciclo for.
xx = np.linspace(0.0,800.0, 31)
yy = catenaria(xx, popt[0], popt[1], popt[2])
# yy = np.array([catenaria(x, popt[0], popt[1], popt[2]) for x in xx])

# Aggiungiamo alla figura precedente la curva di fit.
fig = plt.figure(figsize=(3.0,3.0), frameon=True)
plt.axes([0.25,0.2,0.7,0.7])
plt.plot(px, py, "+r")
plt.plot(xx, yy, "-b")
plt.xlim([0,800])
plt.ylim([-800,0])
plt.xlabel(r"x [px]")
plt.ylabel(r"y [py]")
plt.savefig("fit.png", dpi=300)

# Possiamo anche produrre una figura che mostra all'interno degli assi la
# fotografia stessa che abbiamo usato per determinare le coordinate dei pixel.
# A questo scopo usiamo la funzione "imread" per acquisire la fotografia in
# una variabile ("foto").
# Poi, usiamo la funzione "imshow" per mostrare la figura sugli assi.
# E' necessario definire il parametro "extent" per indicare a imshow dove
# posizionare la foto all'interno degli assi.  I valori sono dati in pixel e 
# corrispondono a (x_min, x_max, y_min, y_max).  L'estensione in pixel lungo
# i due assi deve corrispondere alla dimensione della foto (800x600 nel nostro
# caso).  La figura sarà posizionata a valori negativi dell'asse y perché
# l'asse y ha lo zero posto sul bordo superiore della foto (come discusso
# all'inizio dello script).
foto = matplotlib.image.imread("./catenaria.jpg")
fig = plt.figure(figsize=(3.0,3.0), frameon=True)
plt.axes([0.25,0.2,0.7,0.7])
plt.imshow(foto, extent=(0,800,-600,0))
plt.plot(xx, yy, "--b")
plt.xlim([0,800])
plt.ylim([-800,0])
plt.xlabel(r"x [px]")
plt.ylabel(r"y [py]")
plt.savefig("fit.png", dpi=300)

plt.show()

```
